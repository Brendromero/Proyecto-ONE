### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt


url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación

```python
#Sumo los ingresos de cada tienda

ingreso1 = tienda['Precio'].sum()
ingreso2 = tienda2['Precio'].sum()
ingreso3 = tienda3['Precio'].sum()
ingreso4 = tienda4['Precio'].sum()

print(f'El ingreso de la tienda 1 fue de {ingreso1}')
print(f'El ingreso de la tienda 2 fue de {ingreso2}')
print(f'El ingreso de la tienda 3 fue de {ingreso3}')
print(f'El ingreso de la tienda 4 fue de {ingreso4}')
```

```python
#Creo una tabla resumen

ingresos = pd.DataFrame({"Tienda": ["tienda1", "tienda2", "tienda3", "tienda4"],
                         "Ingreso total": [ingreso1, ingreso2, ingreso3, ingreso4]})

print(ingresos)
```

# 2. Ventas por categoría

```python
#Agrupo por categoria y contar productos vendidos

ventas_t1 = tienda.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_t2 = tienda2.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_t3 = tienda3.groupby('Categoría del Producto').size().sort_values(ascending=False)
ventas_t4 = tienda4.groupby('Categoría del Producto').size().sort_values(ascending=False)

print(f"Las ventas por categoria de la tienda 1 son: \n{ventas_t1}")
```

```python
print(f"Las ventas por categoria de la tienda 2 son: \n{ventas_t2}")
```

```python
print(f"Las ventas por categoria de la tienda 3 son: \n{ventas_t3}")
```

```python
print(f"Las ventas por categoria de la tienda 4 son: \n{ventas_t4}")
```

```python
plt.figure(figsize=(14, 8))

plt.subplot(2, 2, 1)
ventas_t1.plot(kind='bar', color='skyblue')
plt.title("Tienda 1 - Ventas por Categoría")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)

plt.subplot(2, 2, 2)
ventas_t2.plot(kind='bar', color='lightgreen')
plt.title("Tienda 2 - Ventas por Categoría")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)

plt.subplot(2, 2, 3)
ventas_t3.plot(kind='bar', color='salmon')
plt.title("Tienda 3 - Ventas por Categoría")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)

plt.subplot(2, 2, 4)
ventas_t4.plot(kind='bar', color='plum')
plt.title("Tienda 4 - Ventas por Categoría")
plt.ylabel("Cantidad de ventas")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# Calculo la calificacion promedio de cada tienda
promedios = {
    "Tienda 1": tienda["Calificación"].mean(),
    "Tienda 2": tienda2["Calificación"].mean(),
    "Tienda 3": tienda3["Calificación"].mean(),
    "Tienda 4": tienda4["Calificación"].mean(),
}
```

```python
# Lo convierto a DataFrame para mostrar ordenado
promedios_df = pd.DataFrame(list(promedios.items()), columns=["Tienda", "Calificación Promedio"])
print(promedios_df)
```

```python
plt.figure(figsize=(9, 4))
plt.bar(promedios_df["Tienda"], promedios_df["Calificación Promedio"], color=["skyblue", "lightgreen", "salmon", "plum"])
plt.title("Calificación Promedio de Clientes por Tienda")
plt.ylabel("Calificación Promedio")
plt.ylim(0, 5)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

# 7. Informe Final


El objetivo de este informe es sintetizar los hallazgos derivados de los datos de cuatro tiendas y con base en ellos, recomendar a cuál de ellas debería vender el Sr. Juan.
Se evaluaron métricas comerciales y de satisfacción (ingresos, ventas por categoría, calificaciones, productos más y menos vendidos y coste de envío) y se generaron visualizaciones para facilitar la interpretación.

# Resultados principales
 * Ingresos totales: La tienda que obtuvo los mayores ingresos fue Tienda 1, seguida de el resto de las tiendas.
 * Categorías más vendidas: En todas las tiendas, las categorías de Electrónica y Muebles fueron las más populares, mientras que libros, articulos de limpieza e instrumentos musicales tuvieron menor demanda.
 * Calificación promedio: la tienda 2 y la tienda 3 presentan las mejores calificaciones de clientes, indicando una mayor satisfacción.
 * Productos más y menos vendidos: En general, los productos Microondas,cama box y kit de bancas destacan entre los más vendidos. Los artículos como la guitarra electrica, auriculares y refrigerador fueron los menos demandados.
 * Costo de envío promedio: La Tienda 4 tiene el envío más económico, mientras que la Tienda 1 presenta los valores más altos.

# Interpretación general

Aunque la Tienda 1 lidera en ingresos, su calificación promedio y costo de envío son menos competitivos.
Por otro lado, la Tienda 2 tiene excelente valoración de clientes, un buen equilibrio en ingresos y costes de envío razonables, lo que refleja una mejor experiencia general para el comprador.

# Conclusión

Tras analizar los datos, se recomienda que el Sr. Juan venda sus productos en la Tienda 2.
Esta tienda combina buena facturación, altas calificaciones de satisfacción y costos de envío accesibles, factores que aumentan la probabilidad de ventas exitosas y clientes satisfechos.
